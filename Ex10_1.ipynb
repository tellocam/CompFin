```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 101)
u = 1 + 2*np.exp(3*x)
uxx =  18*np.exp(3*x)

h = x[1] - x[0]

uxx_fin_dif = np.zeros((len(x)-1))
uxx_fin_dif2 = np.zeros((len(x)-2))
err1 = []
err2 = []

for i in range(1, len(x)-1):
    uxx_fin_dif[i] = (1/h**2) * (u[i + 1] - 2*u[i] + u[i - 1])
    err1.append(abs(uxx[i]-uxx_fin_dif[i]))
for i in range(2, len(x)-2):
    uxx_fin_dif2[i] = (1/h**2) * (-1/12*u[i+2] + 4/3*u[i+1] - 5/2*u[i] + 4/3*u[i-1] - 1/12*u[i-2])
    err2.append(abs(uxx[i]-uxx_fin_dif[i]))

plt.figure()
plt.plot(uxx, label="analytic")
plt.plot(uxx_fin_dif, marker="^", label="fin_dif1")
plt.plot(uxx_fin_dif2, marker="o", label="fin_dif2")
plt.legend()
plt.show()

plt.figure()
plt.loglog(err1, marker="^")
plt.loglog(err2, marker="o")
plt.show()
```
